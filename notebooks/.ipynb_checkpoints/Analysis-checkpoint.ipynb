# Analysis and plotting

```python
import pandas as pd
import os
%matplotlib notebook
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
from ipywidgets import IntText
import numpy as np
```

## Variables

```python
RESULTS_FOLDER = "../results"
CONFIG_INDEX_NAME = "config_index.csv"
CONFIG_SAVING_NAME = "config.json"
ACCURACY_SAVING_NAME = "accuracies.csv"
```

## Loading set of all registered configs

```python
df_configs = pd.read_csv(os.path.join(RESULTS_FOLDER, CONFIG_INDEX_NAME))
df_configs
```

## Plotting single config

```python
config_idx = 1
STORAGE_FOLDER = os.path.join(RESULTS_FOLDER, str(config_idx))
df_measures = pd.read_csv(os.path.join(STORAGE_FOLDER, ACCURACY_SAVING_NAME))
for what in ['loss', 'r2', 'explained_variance']:
    plt.figure()
    for config_id in df_measures.config_id.unique():
        # make subdataset for config id
        sub_df = df_measures[df_measures.config_id == config_id]

        # plot train test and val
        for mode in ['train', 'test', 'val']:
            # plot training
            plt.plot(sub_df.epoch, sub_df[f"{mode}_{what}"], label=f"{config_id}_{mode}-{what}")

    plt.title(f"{what} plot")
    plt.xlabel("epoch")
    plt.ylabel("score")
    if what == 'r2':
        # plt.ylim(-1,1)
        plt.ylim(-50,1)
    elif what == 'loss':
        plt.ylim(0, 300)
    elif what == 'explained_variance':
        plt.ylim(-10_000, 10_000)
    plt.legend()
    plt.show()
```

```python
def plot_config(ax, config_idx, what):
    # load data
    STORAGE_FOLDER = os.path.join(RESULTS_FOLDER, str(config_idx))
    df_measures = pd.read_csv(os.path.join(STORAGE_FOLDER, ACCURACY_SAVING_NAME))

    # Plot the initial data
    for config_id in df_measures.config_id.unique():
        # make subdataset for config id
        sub_df = df_measures[df_measures.config_id == config_id]

        # plot train test and val
        for mode in ['train', 'test', 'val']:
            # plot training
            ax.plot(sub_df.epoch, sub_df[f"{mode}_{what}"], label=f"{config_id}_{mode}-{what}")
    
    ax.set_title(f"{what} plot")
    ax.set_xlabel("epoch")
    ax.set_ylabel("score")
    ax.legend()
```

```python
what = "r2"

# Create the figure and axis objects
fig, ax = plt.subplots()
plt.subplots_adjust(bottom=0.25)

plot_config(ax, 0, what)

# Create the slider objects
axcolor = 'lightgoldenrodyellow'
ax_a = plt.axes([0.25, 0.1, 0.65, 0.03], facecolor=axcolor)
slider_lower_y = Slider(ax_a, 'y-up', -100, -1, valinit=-1)

config_pick = IntText(
    value=0,
    description="config-idx: ",
    disabled=False
)
def update_config(x):
    ax.cla()
    plot_config(ax, config_pick.value, what)
    
config_pick.observe(update_config , names='value')
display(config_pick)

# Define the update function
def update(val):
    ax.set_ylim(val, 1)
    fig.canvas.draw_idle()

# Register the update function with each slider
slider_lower_y.on_changed(update)
update(slider_lower_y.valinit)

# Show the plot
plt.show()
```
